```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import lagrange
```

```python
#读入数据
np.set_printoptions(suppress=True)
df = pd.read_csv('missing_data.csv',sep = ',',header = None)
pd.set_option('precision', 4)
df
```

```python
#拉格朗日插值函数 s为列向量, n 为插值位置, k 前后数据个数
def ployinterp_column(s,n,k = 5):
    y = s[list(range(n - k)) + list(range(n + 1,n + 1 + k))]
    y = y[y.notnull()] #剔除空值
    return lagrange(y.index,list(y))(n)
```

```python
#获取每列空值所在的index  逐值进行填充
for i in df.columns:
    for j in df[df[i].isnull()].index.tolist():
            df.loc[j,i] = ployinterp_column(df[i],j)
```

```python
#将科学计数法变成保留4位小数
df[1] = df[1].apply(lambda x:'{:.4f}'.format(x))
```
